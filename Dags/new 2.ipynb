```python
import pandas as pd
```

```python
url2='https://raw.githubusercontent.com/data-bootcamp-v4/lessons/refs/heads/main/5_6_eda_inf_stats_tableau/project/files_for_project/df_final_experiment_clients.txt' ## not important
url3='https://raw.githubusercontent.com/data-bootcamp-v4/lessons/refs/heads/main/5_6_eda_inf_stats_tableau/project/files_for_project/df_final_web_data_pt_1.txt'
url4='https://raw.githubusercontent.com/data-bootcamp-v4/lessons/refs/heads/main/5_6_eda_inf_stats_tableau/project/files_for_project/df_final_web_data_pt_2.txt'
```

```python
df_var = pd.read_csv(url2)
df_exp2 = pd.read_csv(url3)
df_exp1 = pd.read_csv(url4)

df_exp = pd.concat([df_exp1, df_exp2])
```

```python
df_var
```

```python
df_exp
```

```python
df = pd.merge(df_var, df_exp, on = 'client_id', how = 'inner')
df.dropna(subset = 'Variation', inplace  = True)
```

```python
df
```

```python
df['confirm'] = df['process_step'].apply(lambda x :1 if x == 'confirm' else 0)
df.head(60)
```

```python
df.confirm.unique()
```

```python
df.groupby('visit_id')['confirm'].sum()
```

```python
df_grouped = pd.DataFrame(df.groupby(['visit_id','Variation'])['confirm'].sum()).reset_index()
```

```python
df.groupby(['visit_id','Variation'])['confirm'].sum()
```

```python
df_grouped.confirm.value_counts()
```

```python
df_grouped['com (y/n)'] = df_grouped.confirm.apply(lambda x: 1 if x > 0 else 0)
```

```python
df_grouped['com (y/n)'].value_counts()
```
